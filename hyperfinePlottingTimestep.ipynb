```python
import json
import numpy as np
import matplotlib.pyplot as plt
import os
import glob

# !./run_hyperfine.sh

# Path to the directory containing your results files
results_dir = 'resultsMain/'  # Update this to the path where your results are stored

# Use glob to match all json files in the results directory
json_files = glob.glob(os.path.join(results_dir, '*.json'))

# Initialize lists to hold means, standard deviations, and command names
means = []
stddevs = []
command_names = []

# Read each JSON file and extract the mean and standard deviation
for file_name in json_files:
    with open(file_name, 'r') as file:
        data = json.load(file)
        # Assuming each file contains results for a single command
        for result in data['results']:
            means.append(result['mean'])
            stddevs.append(result['stddev'])
            # Extract a command name or identifier from the command string
            command_name = os.path.basename(result['command']).split()[0]  # Adjust this as needed
            command_names.append(command_name)
            
# Numeric indices for commands
indices = np.arange(len(means))

# Fit a linear model to the data
a, b = np.polyfit(indices, means, 1)

# Generate the values for the fit line
fit_line = a * indices + b

# Plotting the means with error bars
plt.figure(figsize=(10, 6))
plt.errorbar(command_names, means, yerr=stddevs, fmt='o', ecolor='orangered', capsize=5, linestyle='')
plt.title('Mean Execution Time for Each Command')
plt.xlabel('Command')
plt.ylabel('Mean Execution Time (s)')
plt.xticks(rotation=45, ha='right')  # Rotate the x labels to fit longer commands
plt.tight_layout()  # Adjust layout to fit
plt.plot(indices, fit_line, label=f'Fit Line (y = {a:.2f}x + {b:.2f})', color='blue')

plt.annotate(f'Fit Equation: y = {a:.2f}x + {b:.2f}',
             xy=(0.05, 0.95),
             xycoords='axes fraction',
             horizontalalignment='left',
             verticalalignment='top',
             fontsize=12,
             bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='white'))
plt.grid(True)
plt.show()
```

```python

import json
import matplotlib.pyplot as plt
import os
import glob

# List of paths to your results folders
results_folders = ['resultsMain']

# Initialize a dictionary to hold the data for each folder
folder_data = {}

# Process each results folder
for folder_path in results_folders:
    # Use glob to find all JSON files in the current results folder
    json_files = glob.glob(os.path.join(folder_path, '*.json'))
    means = []  # List to store mean times for the current folder
    command_names = []  # List to store command names for the current folder
    
    # Read each JSON file in the current folder
    for file_name in json_files:
        with open(file_name, 'r') as file:
            data = json.load(file)
            # Assuming each file contains results for a single command
            for result in data['results']:
                means.append(result['mean'])
                # Extract a command name or identifier from the command string
                command_name = os.path.basename(result['command']).split()[0]
                command_names.append(command_name)
    
    # Store the extracted data in the dictionary using the folder name as the key
    folder_name = os.path.basename(folder_path)
    folder_data[folder_name] = {
        'means': means,
        'commands': command_names
    }

# Now, plot the data from each folder
plt.figure(figsize=(12, 8))

for folder_name, data in folder_data.items():
    indices = range(len(data['commands']))  # Convert command names to numerical indices
    plt.plot(indices, data['means'], marker='o', linestyle='-', label=folder_name)

plt.title('Comparison of Mean Execution Times Across Different Folders')
plt.xlabel('Command Index')
plt.ylabel('Mean Execution Time (s)')
plt.legend()
plt.grid(True)
plt.show()
```
