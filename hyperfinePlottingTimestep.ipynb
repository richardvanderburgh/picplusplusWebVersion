```python
# import pandas as pd
# import matplotlib.pyplot as plt
# import os

# # os.system("hyperfine --warmup 3 --runs 20 --export-json results.json 'buildLinux/bin/PIC++Main inputFiles/testInput.json'")
# # !./run_hyperfine.sh

# # Load the JSON data
# with open('results.json') as f:
#     data = pd.read_json(f)

# # Extract the relevant information
# times = data['results'][0]['times']
# mean = data['results'][0]['mean']
# min_time = data['results'][0]['min']
# max_time = data['results'][0]['max']

# # Create the plot
# plt.figure(figsize=(10, 6))
# plt.plot(times, marker='o', linestyle='-', label='Execution Time')
# plt.axhline(y=mean, color='r', linestyle='-', label='Mean')
# plt.axhline(y=min_time, color='g', linestyle='--', label='Min')
# plt.axhline(y=max_time, color='b', linestyle='--', label='Max')
# plt.title('Execution Times for Each Run')
# plt.xlabel('Run Number')
# plt.ylabel('Time (seconds)')
# plt.legend()
# plt.grid(True)
# plt.show()

import json
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import shutil

# Jupyter kernels often start without Homebrew paths, so add common bin dirs
# before looking for hyperfine.
for _bin in ("/opt/homebrew/bin", "/usr/local/bin"):
    if os.path.isdir(_bin) and _bin not in os.environ["PATH"].split(os.pathsep):
        os.environ["PATH"] = _bin + os.pathsep + os.environ["PATH"]

# Regenerate benchmark data only if hyperfine is installed.
if shutil.which("hyperfine") is None:
    print("hyperfine not found on PATH. Install it (e.g. `brew install hyperfine`) to "
          "regenerate benchmarks. Falling back to existing files in results/.")
else:
    os.system("./run_hyperfine.sh inputFiles/benchmarkFiles/timestepDOE")

# Path to the directory containing your results files
results_dir = 'results/'

# Use glob to match all json files in the results directory
json_files = sorted(glob.glob(os.path.join(results_dir, '*.json')))

# Initialize lists to hold means, standard deviations, and command names
means = []
stddevs = []
command_names = []

# Read each JSON file and extract the mean and standard deviation
for file_name in json_files:
    if os.path.getsize(file_name) == 0:
        print(f"Skipping empty results file: {file_name}")
        continue
    try:
        with open(file_name, 'r') as file:
            data = json.load(file)
    except json.JSONDecodeError as exc:
        print(f"Skipping invalid JSON in {file_name}: {exc}")
        continue
    for result in data['results']:
        means.append(result['mean'])
        stddevs.append(result['stddev'])
        # Extract a command name or identifier from the command string
        command_name = os.path.basename(result['command']).split()[0]
        command_names.append(command_name)

if not means:
    raise SystemExit(
        "No benchmark data found in results/. Install hyperfine and run "
        "`./run_hyperfine.sh inputFiles/benchmarkFiles/timestepDOE`, then re-run this cell."
    )

# Numeric indices for commands
indices = np.arange(len(means))

# Fit a linear model to the data
a, b = np.polyfit(indices, means, 1)

# Generate the values for the fit line
fit_line = a * indices + b

# Plotting the means with error bars
plt.figure(figsize=(10, 6))
plt.errorbar(command_names, means, yerr=stddevs, fmt='o', ecolor='orangered', capsize=5, linestyle='')
plt.title('Mean Execution Time for Each Command')
plt.xlabel('Command')
plt.ylabel('Mean Execution Time (s)')
plt.xticks(rotation=45, ha='right')  # Rotate the x labels to fit longer commands
plt.tight_layout()  # Adjust layout to fit
plt.plot(indices, fit_line, label=f'Fit Line (y = {a:.2f}x + {b:.2f})', color='blue')

plt.annotate(f'Fit Equation: y = {a:.2f}x + {b:.2f}',
             xy=(0.05, 0.95),
             xycoords='axes fraction',
             horizontalalignment='left',
             verticalalignment='top',
             fontsize=12,
             bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='white'))
plt.grid(True)
plt.show()
```
